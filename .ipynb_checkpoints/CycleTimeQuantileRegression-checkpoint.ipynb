```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('CycleTimeData.xlsx')
data.head()
```

### Lowess

Set the `frac` parameter to fraction of total points. Here it is .2.

```python
from statsmodels.nonparametric.smoothers_lowess import lowess
x = np.arange(len(data))
y = data['CycleTime']

frac = .2
yp = lowess(y, x, frac=frac)[:, 1]
fig, ax = plt.subplots(dpi=140)
ax.scatter(x, y, marker='.')
ax.plot(x, yp, color='black', lw=4)
ax.set_title(f'Lowess Smoothing with frac {frac}');
```

### Quartile analysis with stacked area plot

Find quartiles from **all** data

```python
bins = y.quantile(q=[0, .25, .5, .75, 1])
bins
```

Cut into bins

```python
s = pd.cut(y, bins, include_lowest=True)
s.head()
```

Count cycle times of every overlapping 14 day period and find percent in above bins.

```python
quantile_count = []
n = 14
for i in range(len(y) - n):
    quantile_count.append(s.iloc[i:i+n].value_counts().sort_index().values)

df = pd.DataFrame(np.row_stack(quantile_count), columns=s.cat.categories) / n
df.head()
```

Smooth the percentages.

```python
dfs = df.rolling(28, center=True, min_periods=14).mean().dropna()
dfs.head()
```

Stacked area chart of cycle times by bin group.

* Between days 100 and 200 there were more cycle times in highest quartile (14-42).
* In last 30 days, very few of lowest quartiles (0-3) and (4-6)

```python
fig, ax = plt.subplots(dpi=144)
ax.stackplot(dfs.index, *dfs.values.T, labels=('0-3', '4-6', '7-13', '14-42'))
ax.legend(bbox_to_anchor=(1, 1), loc='upper left', title='Cycle Times')
ax.set_ylim(0, 1)
```

```python
mod = smf.quantreg('CycleTime ~ Order', data)
res = mod.fit(q=.5)
print(res.summary())
```

```python


quantiles = np.arange(0.2, 0.81, 0.2) # many quantiles between .05 and 0.95, increment by .1
def fit_model(q):
    res = mod.fit(q=q)
    return [q, res.params['Intercept'], res.params['Order']] + \
            res.conf_int().loc['Order'].tolist()

models = [fit_model(x) for x in quantiles]
models = pd.DataFrame(models, columns=['q', 'a', 'b', 'lb', 'ub'])

ols = smf.ols('CycleTime ~ Order', data).fit()
ols_ci = ols.conf_int().loc['Order'].tolist()
ols = dict(a = ols.params['Intercept'],
           b = ols.params['Order'],
           lb = ols_ci[0],
           ub = ols_ci[1])

print(models)
print(ols)
```

```python
x = np.arange(data.Order.min(), data.Order.max(), 50)
get_y = lambda a, b: a + b * x

fig, ax = plt.subplots(figsize=(16, 12))

for i in range(models.shape[0]):
    y = get_y(models.a[i], models.b[i])
    ax.plot(x, y, linestyle='dotted', color='blue')

y = get_y(ols['a'], ols['b'])

ax.plot(x, y, color='red', label='OLS')
ax.scatter(data.Order, data.CycleTime, alpha=.2)
ax.set_xlim((0, 310))
ax.set_ylim((0, 50))
legend = ax.legend()
ax.set_xlabel('Order', fontsize=16)
ax.set_ylabel('CycleTime', fontsize=16);
```
