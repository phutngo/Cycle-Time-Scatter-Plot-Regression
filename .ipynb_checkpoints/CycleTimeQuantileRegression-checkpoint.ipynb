```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('CycleTimeData.xlsx')
```

```python
print(data)
```

```python
data.head()
```

```python
mod = smf.quantreg('CycleTime ~ Order', data)
res = mod.fit(q=.5)
print(res.summary())
```

```python


quantiles = np.arange(0.2, 0.81, 0.2) # many quantiles between .05 and 0.95, increment by .1
def fit_model(q):
    res = mod.fit(q=q)
    return [q, res.params['Intercept'], res.params['Order']] + \
            res.conf_int().loc['Order'].tolist()

models = [fit_model(x) for x in quantiles]
models = pd.DataFrame(models, columns=['q', 'a', 'b', 'lb', 'ub'])

ols = smf.ols('CycleTime ~ Order', data).fit()
ols_ci = ols.conf_int().loc['Order'].tolist()
ols = dict(a = ols.params['Intercept'],
           b = ols.params['Order'],
           lb = ols_ci[0],
           ub = ols_ci[1])

print(models)
print(ols)
```

```python
x = np.arange(data.Order.min(), data.Order.max(), 50)
get_y = lambda a, b: a + b * x

fig, ax = plt.subplots(figsize=(16, 12))

for i in range(models.shape[0]):
    y = get_y(models.a[i], models.b[i])
    ax.plot(x, y, linestyle='dotted', color='blue')

y = get_y(ols['a'], ols['b'])

ax.plot(x, y, color='red', label='OLS')
ax.scatter(data.Order, data.CycleTime, alpha=.2)
ax.set_xlim((0, 310))
ax.set_ylim((0, 50))
legend = ax.legend()
ax.set_xlabel('Order', fontsize=16)
ax.set_ylabel('CycleTime', fontsize=16);
```
